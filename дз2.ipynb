## –†–µ—à–µ–Ω–∏–µ:

–í –∑–∞–¥–∞—á–µ —Ç—Ä–µ–±—É–µ—Ç—Å—è –Ω–∞–π—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ä–µ–¥–∏ 7 –≤—ã–±—Ä–∞–Ω–Ω—ã—Ö –µ–¥–∏–Ω–∏—Ü —Ç–µ—Ö–Ω–∏–∫–∏ –Ω–µ –±–æ–ª–µ–µ –¥–≤—É—Ö –º–æ–≥—É—Ç –±—ã—Ç—å –ø–æ–¥–∫–ª—é—á–µ–Ω—ã –∫ —Å–∏—Å—Ç–µ–º–µ ¬´–£–º–Ω—ã–π –¥–æ–º¬ª.



1.  –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π:
–ü–æ—Å–∫–æ–ª—å–∫—É —Ä–µ—á—å –∏–¥—ë—Ç –æ —Ç–æ–º, —á—Ç–æ –∫–∞–∂–¥—ã–π —ç–∫–∑–µ–º–ø–ª—è—Ä —Ç–µ—Ö–Ω–∏–∫–∏ –æ—Ç–Ω–æ—Å–∏—Ç—Å—è –∫ –æ–¥–Ω–æ–π –∏–∑ –¥–≤—É—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–π (—Å ¬´–£–º–Ω—ã–º –¥–æ–º–æ–º¬ª –∏–ª–∏ –±–µ–∑ –Ω–µ–≥–æ), –∞ –≤—ã–±–æ—Ä –∫–∞–∂–¥–æ–≥–æ —ç–∫–∑–µ–º–ø–ª—è—Ä–∞ —è–≤–ª—è–µ—Ç—Å—è –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–º, –º–æ–¥–µ–ª—å –∑–∞–¥–∞—á–∏ –º–æ–∂–Ω–æ –æ–ø–∏—Å–∞—Ç—å –±–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º.

2.   –í –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–∏:

ùëõ = 7 —á–∏—Å–ª–æ –∏—Å–ø—ã—Ç–∞–Ω–∏–π (–≤—ã–±—Ä–∞–Ω–Ω—ã—Ö —ç–∫–∑–µ–º–ø–ª—è—Ä–æ–≤),

ùëù = 1/4 ‚Äî –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —É—Å–ø–µ—Ö–∞ (—Ç–µ—Ö–Ω–∏–∫–∞ —Å ¬´–£–º–Ω—ã–º –¥–æ–º–æ–º¬ª).

image.png

3.  –ù–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è

–ù—É–∂–Ω–æ –Ω–∞–π—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å–æ–±—ã—Ç–∏—è
ùëã ‚â§ 2, —á—Ç–æ —ç–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–Ω–æ:

P(X‚â§2)=P(X=0)+P(X=1)+P(X=2).

```python
from scipy.stats import binom

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –∑–∞–¥–∞—á–∏
n = 7  # –û–±—â–µ–µ —á–∏—Å–ª–æ —ç–∫–∑–µ–º–ø–ª—è—Ä–æ–≤
p = 1 / 4  # –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å "—É—Å–ø–µ—Ö–∞" (—Ç–µ—Ö–Ω–∏–∫–∞ —Å "–£–º–Ω—ã–º –¥–æ–º–æ–º")

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ P(X <= 2)
# –≠—Ç–æ —Å—É–º–º–∞ P(X = 0), P(X = 1) –∏ P(X = 2)
probability = binom.cdf(2, n, p)

# –û–∫—Ä—É–≥–ª—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –¥–æ —Ç—Ä–µ—Ö –∑–Ω–∞–∫–æ–≤
print(f"–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ä–µ–¥–∏ 7 —ç–∫–∑–µ–º–ø–ª—è—Ä–æ–≤ –Ω–µ –±–æ–ª–µ–µ –¥–≤—É—Ö —Å '–£–º–Ω—ã–º –¥–æ–º–æ–º': {probability:.3f}")
```

**–ó–∞–¥–∞—á–∞ 2** (30 –±–∞–ª–ª–æ–≤)

–î–∞–Ω –∑–∞–∫–æ–Ω —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø—É–±–ª–∏–∫–∞—Ü–∏–π –≤ –∫–æ—Ç–æ—Ä–æ–π –≥—Ä—É–ø–ø–µ —Ç–µ–ª–µ–≥—Ä–∞–º-–∫–∞–Ω–∞–ª–æ–≤:

$$
\begin{array}{|c|c|c|c|c|c|c|}
\hline X \text { (—á–∏—Å–ª–æ –ø—É–±–ª–∏–∫–∞—Ü–∏–π) } & 0 & 1 & 2 & 3 & 4 & 5 \\
\hline \mathrm{P} \text { (–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å) } & ? & 0,3 & 0,3 & 0,2 & 0,1 & 0,1 \\
\hline
\end{array}
$$

–ù–∞–π–¥–∏—Ç–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –¥–ª—è X = 0, –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –∏ –¥–∏—Å–ø–µ—Ä—Å–∏—é. –†–µ—à–∏—Ç–µ –∑–∞–¥–∞—á—É "–≤—Ä—É—á–Ω—É—é" –∏ —Å –ø–æ–º–æ—â—å—é python.

## –†–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏
1. –ù–∞–π–¥–µ–º –ø—Ä–æ–ø—É—â–µ–Ω–Ω—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –¥–ª—è ùëã = 0

–°—É–º–º–∞ –≤—Å–µ—Ö –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –≤ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–∏ –¥–æ–ª–∂–Ω–∞ —Ä–∞–≤–Ω—è—Ç—å—Å—è 1. –û–±–æ–∑–Ω–∞—á–∏–º –ø—Ä–æ–ø—É—â–µ–Ω–Ω—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –∫–∞–∫ P(X=0)=p0. –¢–æ–≥–¥–∞

p0+0,3+0,3+0,2+0,1+0,1=1.

p0 = 1‚àí(0,3+0,3+0,2+0,1+0,1) = 1‚àí1 = 0.

2. –ù–∞–π–¥–µ–º –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ

–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ E(X) —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ:

image.png

–≥–¥–µ Xùëñ ‚Äî –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø—É–±–ª–∏–∫–∞—Ü–∏–π, –∞ P(X = Xi) ‚Äî —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏.

–ü–æ–¥—Å—Ç–∞–≤–∏–º –∑–Ω–∞—á–µ–Ω–∏—è:

E(X) = 0‚ãÖ0+1‚ãÖ0,3+2‚ãÖ0,3+3‚ãÖ0,2+4‚ãÖ0,1+5‚ãÖ0

E(X) = 0+0,3+0,6+0,6+0,4+0,5 = 2,4.

3. –ù–∞–π–¥–µ–º –¥–∏—Å–ø–µ—Ä—Å–∏—é

image.png

(–ø—Ä–∏—à–ª–æ—Å—å —Ä–∞—Å–ø–∏—Å–∞—Ç—å —Ä–µ—à–µ–Ω–∏–µ —á–µ—Ä–µ–∑ Word –¥–ª—è —á–∏—Ç–∞–µ–º–æ—Å—Ç–∏)

```python
import numpy as np

# –î–∞–Ω–Ω—ã–µ –∑–∞–¥–∞—á–∏
X = np.array([0, 1, 2, 3, 4, 5])  # –ó–Ω–∞—á–µ–Ω–∏—è X
P = np.array([0.0, 0.3, 0.3, 0.2, 0.1, 0.1])  # –°–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏

# –ü—Ä–æ–≤–µ—Ä–∏–º —Å—É–º–º—É –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π
assert np.isclose(P.sum(), 1), "–°—É–º–º–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å —Ä–∞–≤–Ω–∞ 1."

# –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ
expected_value = np.sum(X * P)

# –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –∫–≤–∞–¥—Ä–∞—Ç–∞
expected_value_squared = np.sum((X ** 2) * P)

# –î–∏—Å–ø–µ—Ä—Å–∏—è
variance = expected_value_squared - expected_value ** 2

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
print(f"–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ: {expected_value:.2f}")
print(f"–î–∏—Å–ø–µ—Ä—Å–∏—è: {variance:.2f}")
```

**–ó–∞–¥–∞—á–∞ 3** (25 –±–∞–ª–ª–æ–≤)


–ö–æ–º–ø–∞–Ω–∏—è –ø—Ä–∏–æ–±—Ä–µ–ª–∞ 10‚ÄØ000 –Ω–æ–≤—ã—Ö –Ω–æ—É—Ç–±—É–∫–æ–≤. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª–æ–º–∫–∏ –æ–¥–Ω–æ–≥–æ –Ω–æ—É—Ç–±—É–∫–∞ –≤ —Ç–µ—á–µ–Ω–∏–µ –º–µ—Å—è—Ü–∞ —Ä–∞–≤–Ω–∞ 0,0002. –ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç—Å—è, —á—Ç–æ –ø–æ–ª–æ–º–∫–∏ –æ—Ç–¥–µ–ª—å–Ω—ã—Ö –Ω–æ—É—Ç–±—É–∫–æ–≤ –ø—Ä–æ–∏—Å—Ö–æ–¥—è—Ç –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞.

**–í–æ–ø—Ä–æ—Å 1**. –ù–∞–π–¥–∏—Ç–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –∑–∞ –º–µ—Å—è—Ü —Å–ª–æ–º–∞–µ—Ç—Å—è —Ä–æ–≤–Ω–æ 1 –Ω–æ—É—Ç–±—É–∫.
–†–µ—à–µ–Ω–∏–µ –æ–±—ä—è—Å–Ω–∏—Ç–µ. (5 –±–∞–ª–ª–æ–≤)


**–í–æ–ø—Ä–æ—Å 2**. –ö–æ–º–ø–∞–Ω–∏—è –ø–ª–∞–Ω–∏—Ä—É–µ—Ç –≤–≤–µ—Å—Ç–∏ –ø—Ä–æ–≥—Ä–∞–º–º—É —Ä–∞—Å—à–∏—Ä–µ–Ω–Ω–æ–π –≥–∞—Ä–∞–Ω—Ç–∏–∏, –ø–æ –∫–æ—Ç–æ—Ä–æ–π –æ–Ω–∞ –±—É–¥–µ—Ç –±–µ—Å–ø–ª–∞—Ç–Ω–æ —Ä–µ–º–æ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å –∏–ª–∏ –∑–∞–º–µ–Ω—è—Ç—å –Ω–æ—É—Ç–±—É–∫–∏, —Å–ª–æ–º–∞–≤—à–∏–µ—Å—è –≤ —Ç–µ—á–µ–Ω–∏–µ –º–µ—Å—è—Ü–∞. –ï—Å–ª–∏ –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª–æ–º–æ–∫ –∑–∞ –º–µ—Å—è—Ü –ø—Ä–µ–≤—ã—Å–∏—Ç 5 –Ω–æ—É—Ç–±—É–∫–æ–≤, –∫–æ–º–ø–∞–Ω–∏—è –ø–æ–Ω–µ—Å–µ—Ç –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω—ã–µ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ —Ä–∞—Å—Ö–æ–¥—ã.

–∞) –ù–∞–π–¥–∏—Ç–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –∑–∞ –º–µ—Å—è—Ü —Å–ª–æ–º–∞–µ—Ç—Å—è –±–æ–ª–µ–µ 5 –Ω–æ—É—Ç–±—É–∫–æ–≤.–†–µ—à–µ–Ω–∏–µ –æ–±—ä—è—Å–Ω–∏—Ç–µ (10 –±–∞–ª–ª–æ–≤)

–±) –ï—Å–ª–∏ —Å—Ç–æ–∏–º–æ—Å—Ç—å —Ä–µ–º–æ–Ω—Ç–∞ –∏–ª–∏ –∑–∞–º–µ–Ω—ã –æ–¥–Ω–æ–≥–æ –Ω–æ—É—Ç–±—É–∫–∞ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 500 –¥–æ–ª–ª–∞—Ä–æ–≤, –≤—ã—á–∏—Å–ª–∏—Ç–µ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –æ–±—â–∏—Ö —Ä–∞—Å—Ö–æ–¥–æ–≤ –∫–æ–º–ø–∞–Ω–∏–∏ –Ω–∞ –ø–æ–ª–æ–º–∫–∏ –∑–∞ –æ–¥–∏–Ω –º–µ—Å—è—Ü. (10 –±–∞–ª–ª–æ–≤)

```python
import math
from scipy.stats import poisson

# –î–∞–Ω–Ω—ã–µ –∑–∞–¥–∞—á–∏
n = 10000  # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–æ—É—Ç–±—É–∫–æ–≤
p = 0.0002  # –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª–æ–º–∫–∏ –æ–¥–Ω–æ–≥–æ –Ω–æ—É—Ç–±—É–∫–∞
lambda_ = n * p  # –ü–∞—Ä–∞–º–µ—Ç—Ä —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞

# –í–æ–ø—Ä–æ—Å 1: –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å–ª–æ–º–∞–µ—Ç—Å—è —Ä–æ–≤–Ω–æ 1 –Ω–æ—É—Ç–±—É–∫
p_x_1 = poisson.pmf(1, lambda_)
print(f"–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å–ª–æ–º–∞–µ—Ç—Å—è —Ä–æ–≤–Ω–æ 1 –Ω–æ—É—Ç–±—É–∫: {p_x_1:.4f}")

# –í–æ–ø—Ä–æ—Å 2a: –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å–ª–æ–º–∞–µ—Ç—Å—è –±–æ–ª–µ–µ 5 –Ω–æ—É—Ç–±—É–∫–æ–≤
p_x_le_5 = poisson.cdf(5, lambda_)  # P(X <= 5)
p_x_gt_5 = 1 - p_x_le_5  # P(X > 5)
print(f"–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å–ª–æ–º–∞–µ—Ç—Å—è –±–æ–ª–µ–µ 5 –Ω–æ—É—Ç–±—É–∫–æ–≤: {p_x_gt_5:.4f}")

# –í–æ–ø—Ä–æ—Å 2b: –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ —Ä–∞—Å—Ö–æ–¥–æ–≤ –Ω–∞ —Ä–µ–º–æ–Ω—Ç
cost_per_repair = 500  # –°—Ç–æ–∏–º–æ—Å—Ç—å —Ä–µ–º–æ–Ω—Ç–∞ –æ–¥–Ω–æ–≥–æ –Ω–æ—É—Ç–±—É–∫–∞, $
expected_repairs_cost = cost_per_repair * lambda_
print(f"–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ —Ä–∞—Å—Ö–æ–¥–æ–≤ –Ω–∞ —Ä–µ–º–æ–Ω—Ç: ${expected_repairs_cost:.2f}")
```

**–í–æ–ø—Ä–æ—Å 2a. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —Å–ª–æ–º–∞–µ—Ç—Å—è –±–æ–ª–µ–µ 5 –Ω–æ—É—Ç–±—É–∫–æ–≤**
