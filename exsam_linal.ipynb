$$ \vec{a}_2 = \begin{pmatrix} -2; \ 3; \ 3 \end{pmatrix} $$

$$ \vec{a}_3 = \begin{pmatrix} 4; \ -11; \ 9 \end{pmatrix} $$

Матричное представление системы векторов:
формируем матрицу \( A \) из данных векторов:
   \[
   A = \begin{pmatrix}
   1 & -2 & 4 \\
   8 & 3 & -11 \\
   -1 & 3 & 9
   \end{pmatrix}
   \]

**Решение:**

Для определения линейной зависимости/независимости векторов составим определитель матрицы из этих векторов:

$$ \det A = 1 \cdot 3 \cdot 9 + (-2) \cdot (-11) \cdot (-1) + 4 \cdot 8 \cdot 3 - 4 \cdot 3 \cdot (-1) - 1 \cdot (-11) \cdot 3 - (-2) \cdot 8 \cdot 9 $$

$-4 \cdot 3 \cdot (-1) = 12$
$-1 \cdot (-11) \cdot 3 = 33$
$-(-2) \cdot 8 \cdot 9 = 144$

Сложим все члены: $$ \det A = 27 - 22 + 96 + 12 + 33 + 144 = 290 $$

Результат вычислений: $$ \det(A_{3×3}) = 1(4 + 4) - 2(0 - 2) + 3(0 - 4) = 8 + 4 - 12 = 0 $$

**Вывод:**

Так как определитель подматрицы 3×3 равен нулю, система векторов - линейно зависимая при любом значении параметра $a$.

$$ A^T = \begin{pmatrix} -2 & 0 & 1 \ 0 & -2 & 1 \ 1 & 1 & -1 \end{pmatrix} $$

$$ A^TA = \begin{pmatrix} 5 & 1 & -3 \ 1 & 5 & -3 \ -3 & -3 & 3 \end{pmatrix} $$

3. Составляем характеристическое уравнение:
$$ \det(A^TA - λI) = \begin{vmatrix} 5-λ & 1 & -3 \ 1 & 5-λ & -3 \ -3 & -3 & 3-λ \end{vmatrix} = 0 $$

$$ λ_1 = 9, \quad λ_2 = 4, \quad λ_3 = 0 $$

# Задание 10

```python
import numpy as np

# Разделяем данные на предикторы и целевую переменную
X = numeric_columns_filtered.drop(['cnt', 'mnth'], axis=1)
y = numeric_columns_filtered['cnt']

# Добавляем столбец единиц для свободного члена
X = np.column_stack([np.ones(len(X)), X])

# Реализация линейной регрессии
class LinearRegression:
    def fit(self, X, y):
        self.coefficients = np.linalg.inv(X.T @ X) @ X.T @ y
        return self

    def predict(self, X):
        return X @ self.coefficients

# Обучаем модель
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Метрики качества
def r2_score(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    return 1 - (ss_res / ss_tot)

def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Вычисляем метрики
r2 = r2_score(y, y_pred)
mse_value = mse(y, y_pred)

print("Метрики качества модели:")
print(f"R² score: {r2:.3f}")
print(f"MSE: {mse_value:.3f}")
```
